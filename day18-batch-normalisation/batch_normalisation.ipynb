```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate concentric circles data
n_samples = 1000
X, y = make_circles(n_samples=n_samples, factor=0.5, noise=0.05)

# Create a DataFrame
df = pd.DataFrame(np.hstack((X, y.reshape(-1, 1))), columns=['X', 'Y', 'class'])

# Display the first few rows
print(df.head())

# Save the dataset to a CSV file if needed
df.to_csv('concentric_circles.csv', index=False)
```

```python
# Plot the data
plt.scatter(df['X'], df['Y'], c=df['class'])
plt.show()
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,2].values
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
```

```python
model = Sequential()
model.add(Dense(10,activation='relu',input_dim=2))

model.add(Dense(10,activation='relu'))

model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history1 = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
model = Sequential()
model.add(Dense(10,activation='relu',input_dim=2))
model.add(BatchNormalization())

model.add(Dense(10,activation='relu'))
model.add(BatchNormalization())

model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
history2 = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
plt.plot(history1.history['val_accuracy'],color='r')
plt.plot(history2.history['val_accuracy'],color='b')
plt.show()
```
